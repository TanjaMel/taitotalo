```python
import pandas as pd

url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/olympics.csv"

olympics = pd.read_csv(url)


```

```python
print(olympics.shape)
```

```python
# poista tyhjät arvot ja duplikaatit, jos mahdollista:
#tyhjät arvot
olympics = olympics.dropna()

#duplikaatit
olympics = olympics.drop_duplicates()
```

```python
print(olympics.shape)
```

```python
# rajaa tiedot vuodesta 2000 eteenpäin
olympics = olympics[olympics['year'] >= 2000]
```

```python
#rivit ja sarakkeet
print(olympics.shape)
```

visualisoi: 

```python
#tee histogrammi ikäjakaumasta, voi ottaa sukupuolen mukaan

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.histplot(
    data=olympics,
    x="age",
    hue="sex",          
    multiple="stack",    
    bins=20,
    palette="Set1"
)

plt.title("Ikäjakauma sukupuolen mukaan")
plt.xlabel("Ikä")
plt.ylabel("Lukumäärä")
plt.show()
```

```python
#tee boxplot sukupuoli suhteessa mitaleihin
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

sns.boxplot(
    data=olympics,
    x="medal",    
    y="age",       
    hue="sex", 
    palette="Set2"
)

plt.title("Ikäjakauma sukupuolen ja mitalin mukaan (Boxplot)")
plt.xlabel("Mitali")
plt.ylabel("Ikä")
plt.legend(title="Sukupuoli")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#tee pinottu pylväskaavio, jossa x-akselilla plt.title("Mitalien määrä maittain (pinottu pylväskaavio)")

#ryhmän maa + metali
medals_by_country = olympics.groupby(["team", "medal"]).size().unstack(fill_value=0)
#ryhmän maa + metali jarjestuksessa ja tarkistus
medals_by_country = medals_by_country.sort_values(["Gold", "Silver", "Bronze"], ascending=False).head()
#pylväskaavio
medals_by_country.plot(
  kind="bar",
  stacked=True, 
  figsize=(14, 7),
  colormap="Set2"
)

plt.title("Mitalien määrä maittain (pinottu pylväskaavio")
plt.xlabel("Maa")
plt.ylabel("mitalien määrä")
plt.legend(title="Mitali")
plt.tight_layout()
plt.show()
```
